```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.taiwanlottery.com.tw/index_new.aspx'
r = requests.get(url)
#print(r.text)
sp = BeautifulSoup(r.text,'lxml')
#print(sp)
```

```python
#威力彩頭獎預估金額
data_all = sp.find('div',class_ ='top_dollar_tx')
#print(data_all)
datas = sp.find('div',class_ = 'top_dollar').text
print("目前威力彩頭獎預估金額:",datas)
```

```python
#大樂透頭獎預估金額
data_all = sp.find_all('div',class_ ='top_dollar_tx')
#print(data_all)
datas = data_all[1]
data = datas.find('div',class_ = 'top_dollar').text
print("目前大樂透頭獎預估金額:",data)
```

```python
datas = sp.find('div',class_ ='contents_box01')
title = datas.find('span','font_black15').text
print('期數:',title)
nums = datas.find_all('div',class_ = 'ball_tx ball_yellow')
#print(nums)
print("開出獎號:" , end = ' ')
for i in range(20):
    print(nums[i].text , end = " ")
    
```

```python
#猜單雙、猜大小
data_all = sp.find('div',class_ ='contents_box01')
#print(data_all)
data_super = data_all.find('div',class_ ='ball_red').text
print("超級號碼:",data_super)
data_big = data_all.find('div',class_ ="ball_blue_BB1").text
print("猜大小:",data_big)
data_odd = data_all.find('div',class_ = "ball_blue_BB2").text
print("猜單雙:",data_odd)
```

```python
datas = sp.find('div',class_ ='contents_box06')
title = datas.find('span','font_black15').text
print('期數:',title)
nums = datas.find_all('div',class_ = 'ball_tx ball_blue')
#print(nums)
print("開出順序:" , end = ' ')
for i in range(12):
    print(nums[i].text , end = " ")
print("\n大小順序:" , end = ' ')
for i in range(12,24):
    print(nums[i].text , end = " ")
```
