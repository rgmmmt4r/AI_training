```python
import csv
import matplotlib.pyplot as plt
with open("TaipeiWeatherJan.csv",newline = "") as csvfile:
    rows = csv.reader(csvfile)
    Date = []
    HighTemperature = []
    MeanTemperature = []
    LowTemperature = []
    for row in rows:
        #print(row)
        Date.append(row[0][-2:])
        HighTemperature.append(row[1])
        MeanTemperature.append(row[2])
        LowTemperature.append(row[3])
Date = Date[1:]
for i in range(9):
    Date[i] = Date[i][1]
#print(Date)
Hightest = 0
mean = 0
lowest = 100
HighTemperature = HighTemperature[1:]
for temp in HighTemperature:
    if int(temp) > Hightest:
        Hightest = int(temp)
MeanTemperature = MeanTemperature[1:]
for temp in MeanTemperature:
    mean += int(temp)
mean = mean /31
LowTemperature = LowTemperature[1:]
for temp in LowTemperature:
    if int(temp) < lowest:
        lowest = int(temp)
print("最高溫:",Hightest)
print("平均溫:",mean)
print("最低溫:",lowest)
```

```python
plt.plot(Date,HighTemperature)
plt.title("HighTemperature")
plt.xlabel("Date")
plt.ylabel("Temperature")
```

```python
plt.plot(Date,MeanTemperature)
plt.title("MeanTemperature")
plt.xlabel("Date")
plt.ylabel("Temperature")
```

```python
plt.plot(Date,LowTemperature)
plt.title("LowTemperature")
plt.xlabel("Date")
plt.ylabel("Temperature")
```
